```python
#Statistics-Advance-4-Assignment
"""Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test."""
Ans: A t-test is a type of inferential statistic used to determine if there is a significant difference between two 
population means. It is used when the sample size is small and the population standard deviation is unknown. A z-test, 
on the other hand, is a type of inferential statistic used to determine if there is a significant difference between two
population means. It is used when the sample size is large and the population standard deviation is known.

Example scenario for t-test: You are a researcher studying the effectiveness of a new drug on reducing blood pressure. 
You have a sample size of 20 patients and you want to compare the mean blood pressure of the patients before and after 
taking the drug. You would use a t-test to determine if there is a significant difference between the two means.

Example scenario for z-test: You are a researcher studying the effectiveness of a new drug on reducing cholesterol levels.
You have a sample size of 200 patients and you know the population standard deviation. You would use a z-test to 
determine if there is a significant difference between the two means.

"""Q2: Differentiate between one-tailed and two-tailed tests."""
Ans: A one-tailed test is a statistical test in which the critical area of a distribution is one-sided, either greater 
than or less than a certain value. This type of test is used to determine whether an observed result is significantly 
different from the expected result. A two-tailed test, on the other hand, is a statistical test in which the critical 
area of a distribution is two-sided, with values greater than and less than a certain value. This type of test is used 
to determine whether an observed result is significantly different from the expected result in either direction.

"""Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error."""
Ans:
Type 1 Error: A type 1 error is a false positive. This occurs when the null hypothesis is rejected, even though it is 
true. An example of this would be a medical test that incorrectly diagnoses a healthy patient as having a disease. 

Type 2 Error: A type 2 error is a false negative. This occurs when the null hypothesis is accepted, even though it is 
false. An example of this would be a medical test that incorrectly diagnoses a sick patient as being healthy.

"""Q4: Explain Bayes's theorem with an example."""
Ans: Bayes theorem is a mathematical formula used to calculate the probability of an event occurring, given the 
probability of another event that has already occurred. It is named after Thomas Bayes, an 18th-century English 
statistician.

Example: 
Lets say you are trying to determine the probability that a person has a certain disease, given that they have tested 
positive for it.

Using Bayes's theorem, you would calculate the probability of having the disease (P(D)) given a positive test result 
(P(T|D)) multiplied by the prior probability of having the disease (P(D)) divided by the probability of a positive test 
result (P(T)).

Mathematically, this would be expressed as: 
P(D|T) = P(T|D) * P(D) / P(T)

"""Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example."""
Ans: 
A confidence interval is a range of values that is used to estimate a population parameter. It is calculated by taking 
a sample from the population and using the sample statistic to estimate the population parameter. For example, if you 
wanted to estimate the average height of adults in the United States, you could take a sample of adults and calculate 
the average height of the sample. The confidence interval would be the range of values that you are 95% confident 
contains the true population parameter. To calculate the confidence interval, you would need to use a statistical 
formula that takes into account the sample size, standard deviation, and confidence level.

"""Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution."""

Ans: Problem: A patient has a 10% chance of having a certain disease. After taking a test, the patient tests positive 
for the disease. What is the probability that the patient has the disease?

Solution: Using Bayes' Theorem, we can calculate the probability of having the disease given the test result.

P(Disease|Test+) = (P(Test+|Disease) * P(Disease)) / P(Test+)

P(Disease|Test+) = (0.9 * 0.1) / 0.1

P(Disease|Test+) = 0.9

Therefore, the probability that the patient has the disease is 90%.

"""Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5. Interpret the results."""

Ans: The 95% confidence interval is (47.5, 52.5). This means that we are 95% confident that the true population mean 
lies between 47.5 and 52.5.

"""Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error."""
Ans: The margin of error in a confidence interval is the amount of uncertainty associated with a sample statistic. It is
calculated by taking the critical value (determined by the confidence level) multiplied by the standard error of the 
statistic. The sample size affects the margin of error because larger samples tend to produce more precise estimates.

For example, if you are conducting a survey to determine the average age of people in a certain city, a larger sample 
size would result in a smaller margin of error. This is because the larger sample size would provide more accurate 
information about the population, resulting in a more precise estimate of the average age.

"""Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results."""
Ans: z-score = (75 - 70) / 5 = 1

This means that the data point is 1 standard deviation above the population mean.

"""Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test."""

Ans: Null Hypothesis: The drug is not significantly effective at a 95% confidence level. 
Alternative Hypothesis: The drug is significantly effective at a 95% confidence level. 

Test Statistic: t-test 
Significance Level: 0.05 

Calculated t-value: 6/(2.5/sqrt(50)) = 12.24 
Critical t-value: 1.67 

Since the calculated t-value (12.24) is greater than the critical t-value (1.67), we reject the null hypothesis and 
conclude that the drug is significantly effective at a 95% confidence level.

"""Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job."""
Ans: The 95% confidence interval is (0.59, 0.71).

"""Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
significant difference in student performance using a t-test with a significance level of 0.01."""

Ans: Null Hypothesis: There is no significant difference in student performance between the two teaching methods. 

Alternative Hypothesis: There is a significant difference in student performance between the two teaching methods. 

Test Statistic: t = (85-82)/sqrt(6^2/50 + 5^2/50) = 2.45

Degrees of Freedom: df = 50 + 50 - 2 = 98

Critical Value: t-critical (0.01) = 3.36

Since the test statistic (2.45) is less than the critical value (3.36), we fail to reject the null hypothesis. This 
means that there is no significant difference in student performance between the two teaching methods at a significance 
level of 0.01.

"""Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
of 65. Calculate the 90% confidence interval for the true population mean."""
Ans: The 90% confidence interval for the true population mean is (62.4, 67.6).

"""Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average
reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test."""
Ans: 
Null Hypothesis: The caffeine does not have a significant effect on reaction time. 
Alternative Hypothesis: The caffeine has a significant effect on reaction time. 

The t-test statistic is calculated as follows: 
t = (0.25 - 0) / (0.05 / √30) = 10

The critical value for a 90% confidence level is 1.65. Since the t-test statistic is greater than the critical value, 
we reject the null hypothesis and conclude that the caffeine has a significant effect on reaction time at a 90% 
confidence level.

```
